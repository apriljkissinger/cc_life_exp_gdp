# Life Expectancy and GDP

## Introduction

This is a portfolio project for codecademy on life expectancy and GDP. The data is from the World Health Organization and the World Bank. The goal of this project is to try and identify the relationship between life expectancy and GDP in six countries. 

Questions to answer:  
1). How are life expectancy and gdp correlated?  
2). How has life expectancy and gdp changed throughout the years in each country?
 

## Importing Modules

```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

## Loading Data

```python
gdp_life = pd.read_csv('all_data.csv')
print(gdp_life.head())
```

## Exploratory Analysis

Want to look at the unique values of country and year and see the max and min of the life expectancy and GDP data.

```python
gdp_life.Country.unique()
```

```python
gdp_life.Year.unique()
```

Before I find the max and min of the life expectancy, I will rename the column to make it easier to call in the future.

```python
gdp_life.rename( columns ={'Life expectancy at birth (years)': 'Life_Exp'}, inplace=True)
```

```python
print(gdp_life.Life_Exp.min(),gdp_life.Life_Exp.max())
```

```python
print(gdp_life.GDP.min(), gdp_life.GDP.max())
```

There are six countries: Chile, China, Germany, Mexico, United States of America, and Zimbabwe.  
The year spans from 2000-2015.  
The life expectancy from birth spans from 44.3-81.0 years.  
The GDP is in US dollars and spans from approximately \\$4 Billion to $18 Trillion.  

## Data Analysis

To answer the question of how life expectancy is correlated with GDP, first I graph GDP vs life expectancy colorized by country.

```python
ax1 = sns.scatterplot(data=gdp_life, x='Life_Exp', y='GDP', hue='Country')
plt.title('GDP vs Life Expectancy per Country')
ax1.set_xlabel('Life Expectancy')
ax1.set_ylabel('GDP (10s of Trillions))')
```

### Figure 1: GDP vs Life Expectancy per Country

Figure 1 shows GDP vs life expectancy colorized by country. These points show that as GDP increases, so does life expectancy, with the exception of Zimbabwe which seems to show that while the life expectancy is increasing, probably due to infractastructure and medicinal advancements, GDP is not changing. 

What is the trend over the years?

```python
ax2 = sns.scatterplot(data=gdp_life, x='Life_Exp', y='GDP', hue='Year')
plt.title('GDP vs Life Expectancy by Year')
ax2.set_xlabel('Life Expectancy')
ax2.set_ylabel('GDP (10s of Trillions))')
```

### Figure 2: GDP vs Life Expectancy by Year

Figure 2 is colorized by year and it shows that bot life expectancy and GDP are increasing over the years. This is especially true for China and the US as their slopes are the steepest.

Now to build six plots with twin axis of GDP and life expectancy vs year, to compare each country individually. 

```python
plt.figure(figsize=(16, 20))
plt.subplots_adjust(right=1.7)

ax1 = plt.subplot(2, 3, 1)
plt.plot(gdp_life.Year[gdp_life.Country=='Chile'], gdp_life.Life_Exp[gdp_life.Country=='Chile'], color='blue')
ax12 = ax1.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='Chile'], gdp_life.GDP[gdp_life.Country=='Chile'], color='green')
plt.title('Chile', fontsize=20)
ax1.set_xlabel('Year', fontsize=20)
ax1.set_ylabel('Life Expectancy', fontsize=20)
ax12.set_ylabel('GDP (100s of Billions))', fontsize=20)

ax2 = plt.subplot(2, 3, 2)
plt.plot(gdp_life.Year[gdp_life.Country=='China'], gdp_life.Life_Exp[gdp_life.Country=='China'], color='blue')
ax22 = ax2.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='China'], gdp_life.GDP[gdp_life.Country=='China'], color='green')
plt.title('China', fontsize=20)
ax2.set_xlabel('Year', fontsize=20)
ax2.set_ylabel('Life Expectancy', fontsize=20)
ax22.set_ylabel('GDP (10s of Trillions))', fontsize=20)

ax3 = plt.subplot(2, 3, 3)
plt.plot(gdp_life.Year[gdp_life.Country=='Germany'], gdp_life.Life_Exp[gdp_life.Country=='Germany'], color='blue')
ax32 = ax3.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='Germany'], gdp_life.GDP[gdp_life.Country=='Germany'], color='green')
plt.title('Germany', fontsize=20)
ax3.set_xlabel('Year', fontsize=20)
ax3.set_ylabel('Life Expectancy', fontsize=20)
ax32.set_ylabel('GDP (Trillions))', fontsize=20)

ax4 = plt.subplot(2, 3, 4)
plt.plot(gdp_life.Year[gdp_life.Country=='Mexico'], gdp_life.Life_Exp[gdp_life.Country=='Mexico'], color='blue')
ax42 = ax4.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='Mexico'], gdp_life.GDP[gdp_life.Country=='Mexico'], color='green')
plt.title('Mexico', fontsize=20)
ax4.set_xlabel('Year', fontsize=20)
ax4.set_ylabel('Life Expectancy', fontsize=20)
ax42.set_ylabel('GDP (Trillions))', fontsize=20)

ax5 = plt.subplot(2, 3, 5)
plt.plot(gdp_life.Year[gdp_life.Country=='United States of America'], gdp_life.Life_Exp[gdp_life.Country=='United States of America'], color='blue')
ax52 = ax5.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='United States of America'], gdp_life.GDP[gdp_life.Country=='United States of America'], color='green')
plt.title('United States of America', fontsize=20)
ax5.set_xlabel('Year', fontsize=20)
ax5.set_ylabel('Life Expectancy', fontsize=20)
ax52.set_ylabel('GDP (10s of Trillions))', fontsize=20)

ax6 = plt.subplot(2, 3, 6)
plt.plot(gdp_life.Year[gdp_life.Country=='Zimbabwe'], gdp_life.Life_Exp[gdp_life.Country=='Zimbabwe'], color='blue')
ax62 = ax6.twinx()
plt.plot(gdp_life.Year[gdp_life.Country=='Zimbabwe'], gdp_life.GDP[gdp_life.Country=='Zimbabwe'], color='green')
plt.title('Zimbabwe', fontsize=20)
ax6.set_xlabel('Year', fontsize=20)
ax6.set_ylabel('Life Expectancy', fontsize=20)
ax62.set_ylabel('GDP (10s of Billions))', fontsize=20)

#fig.tight_layout()
```
